# using Tips

```python
df=sns.load_dataset('tips')
print(df.head())
```

```python
hue_colors={"Yes": "red", "No":"black"}
sns.scatterplot(x="total_bill", y="tip", 
                data=df, 
                hue="smoker",
                hue_order=["Yes", "No"],
                palette=hue_colors)
plt.show()
```

```python
sns.countplot(x="smoker",
              data=df, 
              hue="sex")

plt.show()
```

# Relational PLots & Sub plots

```python
sns.relplot(x="total_bill", 
            y="tip", 
            data=df,
            kind="scatter",
            col="smoker", # can also row
            row="time"
            )
plt.show()
```

```python
sns.relplot(x="total_bill", 
            y="tip", 
            data=df,
            kind="scatter",
            col="smoker", # can also row
            row="day"
            )
plt.show()
```

```python
sns.relplot(x="total_bill", 
            y="tip", 
            data=df,
            kind="scatter",
            col="day",
            col_order=["Thur","Fri","Sat","Sun"]
            )
plt.show()
```

```python
sns.relplot(x="total_bill", 
            y="tip", 
            data=df,
            kind="scatter",
            size="size",
            hue="size",
            alpha = 0.6
            )
plt.show()
```


# Categorical Plots- Count Plots, bar plots

```python
sns.countplot(
    x="smoker",
    data = df
    )
plt.show()
```

```python
sns.catplot(
    x="smoker",
    data = df,
    kind="count",
    order = ["No", "Yes"]
    )
plt.show()
```

```python
sns.catplot(
    x="day",
    y="total_bill",
    data = df,
    kind="bar",

    )
plt.show()
#short bar uncertanity is low
```

```python
sns.catplot(
    x="total_bill",
    y="day",
    data = df,
    kind="bar"

    )
plt.show()
```

Box Plots

- Shows the distribution of quantative data

- see median, spread, skewness and outliners

```python
sns.catplot(
    x="day",
    y="total_bill",
    data = df,
    kind="box",
   

    )
plt.show()
```

```python
sns.catplot(
    x="time",
    y="total_bill",
    data = df,
    kind="box",
    order = ["Dinner", "Lunch"]
   

    )
plt.show()
```

```python
sns.catplot(
    x="day",
    y="total_bill",
    data = df,
    kind="box",
    sym=""  # Remove the ouliners

    )
plt.show()
```

```python
sns.catplot(
    x="day",
    y="total_bill",
    data = df,
    kind="box",
    sym="",  # Remove the ouliners
    whis=[0,100]
    )
plt.show()
# whis=2.0 make them extend 2.0*IQR
# show the 5th and 95th percentile : whis=[5, 95]
# show min and max values: whis=[0,100]
```

# Bokeh

```python
from bokeh.plotting import figure,output_file,show 
x=[1,2,3,4,5]
y=[2,4,6,8,10]
output_file('line.html')

fig=figure(title="Line Plot",
           x_axis_label="x axis",
           y_axis_label="y axis")

fig.line(x,y)
show(fig)
```
